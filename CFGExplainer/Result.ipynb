```python
import matplotlib.pyplot as plt

# 準備要繪製的資料
x = [i for i in range(14) if i != 8]
y = [0.0220, 0.0679, 0.3740, 0.1100, 0.0860, 0.2940, 0.2020, 0.3080, 0.2260, 0.5480, 0.8080, 0.9460, 0.9540]
z = [0.0220, 0.6600, 0.8420, 0.8460, 0.5859, 0.8800, 0.8300, 0.8580, 0.7479, 0.7680, 0.8700, 0.8540, 0.8739]
# 8: 0.9140

# 創建圖形和軸
fig, ax = plt.subplots(figsize=(8, 6))  # 設置圖的大小，這裡為寬度8英寸，高度6英寸

# 繪製折線圖
ax.plot(x, y, z)

# 添加標籤和標題
ax.set_xlabel('Number of Subgraphs as Trigger')
ax.set_ylabel('Attack Success Rate')
ax.set_title('Relationship between Number of Subgraphs and Attack Success Rate')

# 顯示圖形
plt.show()
```
